```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://hi.wikipedia.org/w/index.php?title=%E0%A4%B5%E0%A4%BF%E0%A4%B6%E0%A5%87%E0%A4%B7:%E0%A4%B8%E0%A4%AD%E0%A5%80_%E0%A4%AA%E0%A5%83%E0%A4%B7%E0%A5%8D%E0%A4%A0&from=1881+%E0%A4%88%E0%A4%B8%E0%A4%BE+%E0%A4%AA%E0%A5%82%E0%A4%B0%E0%A5%8D%E0%A4%B5')
r.encoding = 'utf-8'
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
results = soup.find_all('li', attrs={'class':'allpagesredirect'})
```
